### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
# Shape of the dataset
print("Shape:", df.shape)

# Column names and data types
print("\nData Types:")
print(df.dtypes)

# Summary information
print("\nDataset Info:")
df.info()

# Statistical summary
print("\nStatistical Summary:")
df.describe(include='all')
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Drop columns with > 50% missing
threshold = 0.5
df = df[df.columns[df.isnull().mean() < threshold]]

# Fill remaining missing values (pandas 2.x compatible)
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0]) # mode for categorical
    elif pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(df[col].median()) # median for numeric

print("\nMissing values after handling:")
print(df.isnull().sum().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
# Use only columns that actually exist in df
desired_cols = ['Employment', 'JobSat', 'YearsCodePro']
key_cols = [col for col in desired_cols if col in df.columns]

print("Analyzing columns:", key_cols)

for col in key_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts())
    print(f"Unique values: {df[col].nunique()}")
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Auto-detect the JobSat column
jobsat_col = next((c for c in df.columns if 'sat' in c.lower()), None)

if jobsat_col is None:
    print("No satisfaction column found. Available columns:")
    print(df.columns.tolist())
else:
    print(f"Using column: {jobsat_col}")
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Pie chart
    jobsat_counts = df[jobsat_col].value_counts()
    axes[0].pie(jobsat_counts, labels=jobsat_counts.index,
                autopct='%1.1f%%', startangle=140,
                colors=sns.color_palette('viridis', len(jobsat_counts)))
    axes[0].set_title('Job Satisfaction Distribution (Pie Chart)')

    # Count plot
    sns.countplot(data=df, y=jobsat_col,
                  order=df[jobsat_col].value_counts().index,
                  palette='viridis', ax=axes[1])
    axes[1].set_title('Job Satisfaction Distribution')

    plt.tight_layout()
    plt.show()

    print("\nKey Trends:")
    print(jobsat_counts.head(3))
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# Explode both language columns
def get_lang_counts(df, col):
    return (df[col].dropna()
            .str.split(';')
            .explode()
            .str.strip()
            .value_counts())

worked_counts = get_lang_counts(df, 'LanguageHaveWorkedWith')
want_counts = get_lang_counts(df, 'LanguageWantToWorkWith')

# Top 15 from each
top_n = 15
compare_df = pd.DataFrame({
    'HaveWorkedWith': worked_counts.head(top_n),
    'WantToWorkWith': want_counts.head(top_n)
}).fillna(0)

# Grouped bar chart
compare_df.plot(kind='bar', figsize=(14, 6), colormap='Set1', edgecolor='black')
plt.title('Programming Languages: Have Worked With vs Want to Work With')
plt.xlabel('Language')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
# Define regions
region_map = {
    'United States of America': 'North America', 'Canada': 'North America',
    'United Kingdom of Great Britain and Northern Ireland': 'Europe',
    'Germany': 'Europe', 'France': 'Europe', 'Netherlands': 'Europe',
    'India': 'Asia', 'China': 'Asia', 'Japan': 'Asia',
    'Brazil': 'South America', 'Australia': 'Oceania',
    'South Africa': 'Africa', 'Nigeria': 'Africa'
}
df['Region'] = df['Country'].map(region_map).fillna('Other')

# Cross-tab RemoteWork by Region
remote_region = pd.crosstab(df['Region'], df['RemoteWork'])

# Grouped bar chart
remote_region.plot(kind='bar', figsize=(12, 6), colormap='tab10', edgecolor='black')
plt.title('Remote Work Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=30, ha='right')
plt.legend(title='Remote Work', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
import numpy as np

# Manually set the correct column name
# Common possibilities - uncomment the one that exists in your df:
jobsat_col = 'JobSatPoints_1'
# jobsat_col = 'JobSatisfaction'  
# jobsat_col = 'job_sat'

# Verify it exists
if jobsat_col not in df.columns:
    print("Column not found! Available columns with numbers or 'job':")
    print([c for c in df.columns if any(x in c.lower() for x in ['job', 'sat', 'point'])])
else:
    df['YearsCodePro_num'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
    df['JobSat_num'] = pd.to_numeric(df[jobsat_col], errors='coerce')
    
    clean = df[['YearsCodePro_num', 'JobSat_num']].dropna()
    
    pearson_r = clean['YearsCodePro_num'].corr(clean['JobSat_num'], method='pearson')
    spearman_r = clean['YearsCodePro_num'].corr(clean['JobSat_num'], method='spearman')
    
    print(f"Pearson r: {pearson_r:.3f}")
    print(f"Spearman r: {spearman_r:.3f}")
    
    plt.figure(figsize=(10, 6))
    plt.scatter(clean['YearsCodePro_num'], clean['JobSat_num'], alpha=0.3, color='steelblue')
    m, b = np.polyfit(clean['YearsCodePro_num'], clean['JobSat_num'], 1)
    x_line = np.linspace(clean['YearsCodePro_num'].min(), clean['YearsCodePro_num'].max(), 100)
    plt.plot(x_line, m * x_line + b, color='red', linewidth=2)
    plt.title(f'Experience vs Job Satisfaction (Pearson r={pearson_r:.3f})')
    plt.xlabel('Years of Professional Coding')
    plt.ylabel('Job Satisfaction')
    plt.tight_layout()
    plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Cross-tabulation
cross_tab = pd.crosstab(df['EdLevel'], df['Employment'])
print(cross_tab)

# Stacked bar chart
cross_tab.plot(kind='bar', stacked=True, figsize=(14, 7),
               colormap='tab20', edgecolor='black')
plt.title('Employment Type by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=40, ha='right')
plt.legend(title='Employment', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
