<h3>Step 2: Plot the Distribution of Industry</h3>


Explore how respondents are distributed across different industries.

- Plot a bar chart to visualize the distribution of respondents by industry.

- Highlight any notable trends.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Value counts for industry
industry_counts = df['Industry'].value_counts()

plt.figure(figsize=(12, 6))
industry_counts.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Distribution of Respondents by Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("Top 5 Industries:")
print(industry_counts.head())
```

<h3>Step 3: Identify High Compensation Outliers</h3>


Identify respondents with extremely high yearly compensation.

- Calculate basic statistics (mean, median, and standard deviation) for `ConvertedCompYearly`.

- Identify compensation values exceeding a defined threshold (e.g., 3 standard deviations above the mean).


```python
comp = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce').dropna()

mean = comp.mean()
median = comp.median()
std = comp.std()

print(f"Mean: ${mean:,.2f}")
print(f"Median: ${median:,.2f}")
print(f"Std Dev: ${std:,.2f}")

# Threshold: 3 standard deviations above mean
threshold = mean + 3 * std
high_outliers = df[pd.to_numeric(df['ConvertedCompYearly'], errors='coerce') > threshold]

print(f"\nThreshold (mean + 3×std): ${threshold:,.2f}")
print(f"Number of high outliers: {len(high_outliers)}")
```

<h3>Step 4: Detect Outliers in Compensation</h3>


Identify outliers in the `ConvertedCompYearly` column using the IQR method.

- Calculate the Interquartile Range (IQR).

- Determine the upper and lower bounds for outliers.

- Count and visualize outliers using a box plot.


```python
comp_col = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

Q1 = comp_col.quantile(0.25)
Q3 = comp_col.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1: ${Q1:,.2f}")
print(f"Q3: ${Q3:,.2f}")
print(f"IQR: ${IQR:,.2f}")
print(f"Lower bound: ${lower_bound:,.2f}")
print(f"Upper bound: ${upper_bound:,.2f}")

outliers = df[(comp_col < lower_bound) | (comp_col > upper_bound)]
print(f"\nTotal outliers detected: {len(outliers)}")

# Box plot
plt.figure(figsize=(8, 5))
plt.boxplot(comp_col.dropna(), vert=False, patch_artist=True,
            boxprops=dict(facecolor='steelblue', color='black'))
plt.title('Compensation Distribution (Box Plot)')
plt.xlabel('ConvertedCompYearly (USD)')
plt.tight_layout()
plt.show()
```

<h3>Step 5: Remove Outliers and Create a New DataFrame</h3>


Remove outliers from the dataset.

- Create a new DataFrame excluding rows with outliers in `ConvertedCompYearly`.
- Validate the size of the new DataFrame.


```python
comp_col = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Keep only rows within IQR bounds
df_clean = df[(comp_col >= lower_bound) & (comp_col <= upper_bound)].copy()

print(f"Original DataFrame size: {df.shape}")
print(f"Cleaned DataFrame size: {df_clean.shape}")
print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")
```
